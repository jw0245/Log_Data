## replace()가 dataframe에 적용 안되는 현상



```python
df['Time'] = df.Time.str.replace('[', '', regex=True)
df['Time']
```

```python
# pandas 에서 지원해주는 datetime 형식으로 변환하는 법
df['Time'] = pd.to_datetime(df['Time'], format = '%d/%b/%Y:%H:%M:%S', errors='coerce')
df['Time']
```
